# 데이터 불러와서 Concat

```python
import pandas as pd 
import numpy as np

#데이터 불러오기
df16 = pd.read_sas('hn16_all.sas7bdat')
df17 = pd.read_sas('hn17_all.sas7bdat')
df18 = pd.read_sas('hn18_all.sas7bdat')
```

```python
df_all = pd.concat([df16,df17,df18])
```

```python
df_att = df_all[["sex","age","D_1_1","Total_slp_wd","BP1","BO1","BE5_1","BE8_1","BE8_2","BP5","HE_ht","HE_wt"]]
df_att
```

# 설문조사 데이터 => 8,9 무응답 NA 처리
## info Dtype 변경

```python
#결측치 확인
df_att.info()
```

```python
# BMI [BMI] (명목) 변수 생성
df_att['BMI'] = (df_att['HE_wt'] / (df_att['HE_ht'] / 100)**2)
df_att.loc[df_att['BMI'] < 25, 'BMI'] = 0
df_att.loc[df_att['BMI'] >= 25, 'BMI'] = 1
print(df_att['BMI'].value_counts())
print("==========================")
print(df_att['BMI'])
```

```python
df_att = df_att.drop(['HE_wt','HE_ht'], axis=1)
```

```python
# 하루 평균 앉아서 보내는 시간(분 단위) 변수 생성
df_att["SitTime"] = df_att["BE8_1"]*60 + df_att['BE8_2']
```

```python
df_att = df_att.drop(['BE8_1', 'BE8_2'],axis=1)
```

## 종속변수 float -> 카테고리(순서X) 명목형 

```python
df_att['BMI'] = df_att['BMI'].astype('category')
df_att.info()
```

## 변수 float -> 카테고리화 (순서)

```python
# 주관적 건강상태 [D_1_1]] (순서)
from pandas.api.types import CategoricalDtype
print(df_att['D_1_1'].value_counts())
print("==========================")
df_att.loc[df_att['D_1_1'] == 9, 'D_1_1'] = np.nan
df_att['D_1_1'] = df_att['D_1_1'].astype(CategoricalDtype(ordered=True))
print(df_att['D_1_1'].value_counts())
print("==========================")
print(df_att['D_1_1'])
```

```python
# 주관적 체형 인지 [BO1] (순서)
from pandas.api.types import CategoricalDtype
print(df_att['BO1'].value_counts())
print("==========================")
df_att.loc[df_att['BO1'] == 8, 'BO1'] = np.nan
df_att.loc[df_att['BO1'] == 9, 'BO1'] = np.nan

df_att['BO1'] = df_att['BO1'].astype(CategoricalDtype(ordered=True))
print(df_att['BO1'].value_counts())
print("==========================")
print(df_att['BO1'])
```

```python
df_att.info()
```

```python
# 1주일간 근력운동 일수 (등간)
print(df_att['BE5_1'].value_counts()) # 8 비해당 929명, 9 모름, 무응답 : 267명 -> NA
df_att.loc[(df_att['BE5_1'] == 8) |(df_att['BE5_1'] == 9) , 'BE5_1'] = np.nan  #na 값 처리
print(df_att['BE5_1'].value_counts())
#df_att['BE5_1'].astype('int64') # na 제거하거나 대체하면 에러 안남. 
```

# 청소년 데이터 loc

```python
# 청소년 데이터 생성
df_csn = df_att.loc[(df_att['age']>=12) & (df_att['age']<=18), :]  # 청소년 (만12세~18세) 데이터 생성
df_csn.info()
```

## 1. 결측치 제거

```python
df_delete = df_csn.copy()
```

```python
df_delete = df_delete.dropna(axis=0)
df_delete = df_delete.reset_index()
df_delete
```

### 연속형 변수 outlier 제거

```python
def remove_outlier(d_cp, column):
    fraud_column_data = d_cp[column]
    quan_25 = np.percentile(fraud_column_data, 25)
    quan_75 = np.percentile(fraud_column_data, 75)
    
    iqr = quan_75 - quan_25
    iqr = iqr * 1.5
    lowest = quan_25 - iqr
    highest = quan_75 + iqr
    outlier_index = fraud_column_data[(fraud_column_data<lowest) | (fraud_column_data > highest)].index
    d_cp.drop(outlier_index, axis=0, inplace=True)
    return d_cp
```

```python
df_remove = remove_outlier(df_delete, "SitTime")
df_remove
```

```python
df_remove.isnull().sum()
```

## 연속형 변수 - 표준화, minmax scaling

```python
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler()
standard = transformer.fit_transform(df_remove.loc[:,['Total_slp_wd','SitTime']])
df_standard = pd.DataFrame(standard, columns = ['Total_slp_wd_standard','SitTime_standard'])
df_standard = pd.concat([df_remove, df_standard], axis = 1)
df_standard = df_standard.iloc[:,1:]
df_standard
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
scaled = min_max_scaler.fit_transform(df_remove.loc[:,['Total_slp_wd','SitTime']])
df_scaled = pd.DataFrame(scaled, columns = ['Total_slp_wd_scaled','SitTime_scaled'])
df_all = pd.concat([df_standard, df_scaled], axis=1)
df_all = df_all.iloc[:,0:]
df_all
```

```python
df_all = pd.concat([df_standard, df_scaled], axis=1)
df_all = df_all.iloc[:,1:]
df_all
```

```python
df_all.iloc[:,9:] = round(df_all.iloc[:,9:],2)
```
